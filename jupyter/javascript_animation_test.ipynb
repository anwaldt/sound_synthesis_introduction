```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation

n = 100
wf = np.zeros((n,2))

def next_wf():
    global wf
    offset = wf[0,0] + 0.1
    wf[:,0] = np.linspace(offset, np.pi*4+offset, wf.shape[0])
    wf[:,1] = np.sin(wf[:,0])

def animate(frame):
    next_wf()
    plt.cla()
    plot, = plt.plot(wf[:,0], wf[:,1])
    return plot,

next_wf()
fig, ax = plt.subplots(1)
anim = animation.FuncAnimation(fig, animate, frames=100,interval=25)

HTML(anim.to_jshtml());
```
