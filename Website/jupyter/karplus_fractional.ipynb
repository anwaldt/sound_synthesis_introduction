```python
import math as math

b = np.random.standard_normal(L)

buffer_pos = 0

step = 0.25

y = []

for i in range(0,2*fs):
   
    lp   = math.floor(buffer_pos)
    up   = lp+1
    
    if(up)>=L:
        up=0
        
    frac = buffer_pos%1
    
    val = (b[lp]*(1-frac)) + (b[up]*(frac)) 
    
    y.append(val)
    
    b[lp] = val
    
    # increment buffer counter and wrap it
    buffer_pos = buffer_pos+step
    
    if buffer_pos>=L:
        buffer_pos=0
        
ipd.display(ipd.Audio(y, rate=fs))         
```
