
Waveshaping is one of the basic ways of distortion synthesis. In its simplest form it works like any overdrive effect by limiting a signal with a non-linear shaping function. Depending on the implementation, these shaping
functions can have any form. 

# Shaping Function


The following example shows a simple tangential shaping function
$y=\mathrm{tanh}(g x)$. For high pre-gain values, the function converges towards a step function and the output of a sinusoidal imput signal becomes a square wave.


```python
%run header.ipynb

x    = np.linspace(-1,1,100)
y    = np.tanh(x*4)

p = plt.plot(x,y,'k')

plt.xlabel('Input');
plt.ylabel('Output');
```
